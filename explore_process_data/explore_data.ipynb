# Preprocessing data in the training csv file

Simplestly, we fill all the NaN in the label column as 0.0 as negative. And then drop the administrative columns.

```python
# first we drop the administrative columns
drop_columns = ["Sex", "Age", "AP/PA"]
train_csv = train_csv.drop(columns=drop_columns)
```

```python
train_csv
```

```python
# fill all the NaN as 0.0
train_csv = train_csv.fillna(0)
```

```python
train_csv
```

```python
# write train.csv into the file
file_path = "/scratch/hw1666/final_projects/dl4med/data//CheXpert-v1.0-small"
processed_train_csv_filename = "/train_preprocessed.csv"
train_csv.to_csv(file_path+processed_train_csv_filename, index=False)
```
