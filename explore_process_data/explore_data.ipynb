# Preprocessing data in the training csv file

Simplestly, we fill all the NaN in the label column as 0.0 as negative. And then drop the administrative columns.

```python
# first we drop the administrative columns
drop_columns = ["Sex", "Age", "AP/PA"]
train_csv = train_csv.drop(columns=drop_columns)
```

```python
train_csv
```

```python
# fill all the NaN as 0.0
train_csv = train_csv.fillna(0)
```

```python
train_csv
```

```python
# write train.csv into the file
file_path = "/scratch/hw1666/final_projects/dl4med/data/processed_data"
processed_train_csv_filename = "/train_drop_fill.csv"
train_csv.to_csv(file_path+processed_train_csv_filename, index=False)
```

# Read in the processed train.csv and divide it into two sets

since there is no test set in the original set, so here, we divide the whole training set into two parts: (1) training set (2) validation set. Meanwhile, we use the original validation set as the test set.

```python
# read in the processed train.csv
with open(file_path+processed_train_csv_filename, "rt") as fin:
    train_csv_processed = pd.read_csv(fin)
```

```python
train_csv_processed
```

### divide the training set into training and validation set, note that there are no overlapping of patients between different sets.

Here, we make the validation size as that of original validation set. The intuition is that original paper thinks that the original validation size is enough for tuning the model, so we use the same size, while not decreasing the training set aggressively.

```python
train_csv_processed[-234:]
```

```python
train_csv_processed.iloc[-235]
```

So we can see that there is no overlapping between two sets.

```python
valid_splitted = train_csv_processed[-234:]
train_splitted = train_csv_processed[:-234]
```

```python
train_splitted
```

```python
valid_splitted
```

```python
# get the unique values for each columns in train_splitted dataset
for i, col in enumerate(train_splitted.columns):
    print(i)
    print(col)
    print(pd.unique(train_splitted[col]))
    print("\n")
```

```python
# get the unique values for each columns in train_splitted dataset
for i, col in enumerate(valid_splitted.columns):
    print(i)
    print(col)
    print(pd.unique(valid_splitted[col]))
    print("\n")
```

### shuffle them

```python
train_splitted_shuffled = train_splitted.sample(frac=1).reset_index(drop=True)
```

```python
valid_splitted_shuffled = valid_splitted.sample(frac=1).reset_index(drop=True)
```

```python
train_splitted_shuffled
```

```python
valid_splitted_shuffled
```

```python
# write them into the files
file_path = "/scratch/hw1666/final_projects/dl4med/data/processed_data"
train_filename = "/train_splitted_shuffle.csv"
valid_filename = "/valid_splitted_shuffel.csv"
train_splitted_shuffled.to_csv(file_path+train_filename, index=None)
valid_splitted_shuffled.to_csv(file_path+valid_filename, index=None)
```
